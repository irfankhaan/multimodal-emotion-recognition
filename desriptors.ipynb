```python
import os
path = 'emotional database/descriptors/'
```

```python
files = []
for r, d, f in os.walk(path):
    for file in f:
        if '.txt' in file:
            files.append(os.path.join(r, file))
```

```python
len(files)
```

```python
descriptors = {}
index = 1
for i in files:
    temp = {}
    temp['gender'] = i[38]
    temp['gender_no'] = i[39]
    temp['output'] = i[41:43]
    temp['expression_index'] = i[43]
    file = open(i, 'r')
    s = file.read().split('\n')
    temp['video_name'] = s[1]
    temp['start_time_in_sec'] = s[2].split(':')[0]
    temp['end_time_in_sec'] = s[3].split(':')[0]
    temp['start_exp'] = s[2].split(':')[1]
    temp['end_exp'] = s[3].split(':')[1]
    temp['text_file'] = s[5]
    temp['kinect_start_frame'] = s[6]
    temp['kinect_end_frame'] = s[7]
    descriptors[index] = temp
    index += 1
```

```python
descriptors
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(descriptors, orient='index')
```
